```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

### Read and load the file

```python
dataset=pd.read_csv("flower_species.csv")
```

### Plot the data points on the graph

```python
sns.scatterplot(x=dataset.Sepal_length,y=dataset.Sepal_width,hue=dataset.Species)
sns.set_style("darkgrid")
plt.title("Species Clsassification");
```

### Create a temporary dataframe for sorting

```python
temp_df=dataset.copy()
display(temp_df)
```

### Distance function to  calculate the distance and sort the dataframe by distance and store it in another dataframe

## K=3

```python
def distance(sl,sw):
    k=3
    for i in range(dataset.shape[0]):
        temp_df['Distance'][i]=math.sqrt((temp_df.Sepal_length[i]-sl)**2+(temp_df.Sepal_width[i]-sw)**2)
    sorted_df=temp_df.sort_values('Distance').round(3).reset_index(drop=True)
    print("Sorted Distance Values:\n")
    display(sorted_df)    ###sorted dataframe
    print("The ",k," nearest neighbours of the test data are :\n")
    for i in range(k):
        print(i+1,".",sorted_df.Distance[i])
    k1=sorted_df.Species[0]
    df2=pd.DataFrame({'Sepal_length':[sl],'Sepal_width':[sw],'Species':[k1]})
   
    display( dataset.append(df2, ignore_index = True)) ###final dataframe with the output value in the last row
  
```

# Get the input values from the user
# OUTPUT IS IN THE LAST ROW OF THE DATAFRAME
### Output says which species the flower's input values belongs to.

```python
test_sl=float(input("Enter the sepal length of the flower: "))
test_sw=float(input("Enter the sepal width of the flower: "))
distance(test_sl,test_sw)
```
