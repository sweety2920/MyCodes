### Frequency table for Blood Pressure

```python
def bp_freq(a):
    if a=="high":
            high=data[data['Blood Pressure']=="high" ]
            high_yes=high[high['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            high_no=high[high['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return high_yes,high_no
    if a=="normal":
            normal=data[data['Blood Pressure']=="normal" ]
            normal_yes=normal[normal['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            normal_no=normal[normal['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return normal_yes,normal_no
    if a=="low":
            low=data[data['Blood Pressure']=="low" ]
            low_yes=low[low['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            low_no=low[low['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return low_yes,low_no
```

### Frequency table for Fever

```python
def fever_freq(o):
    if o=="high":
            more=data[data.Fever=="high" ]
            more_yes=more[more['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            more_no=more[more['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return more_yes,more_no
    if o=="mild":
            mild=data[data.Fever=="mild" ]
            mild_yes=mild[mild['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            mild_no=mild[mild['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return mild_yes,mild_no
    if o=="no fever":
            nofever=data[data.Fever=="no fever" ]
            nofever_yes=nofever[nofever['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            nofever_no=nofever[nofever['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return nofever_yes,nofever_no
```

### Frequency table for Diabetes

```python
def diabetes_freq(b):
    if b=="yes":
            dbyes=data[data.Diabetes=="yes" ]
            dbyes_yes=dbyes[dbyes['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            dbyes_no=dbyes[dbyes['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return dbyes_yes,dbyes_no
    if b=="no":
            dbno=data[data.Diabetes=="no" ]
            dbno_yes=dbno[dbno['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            dbno_no=dbno[dbno['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return dbno_yes,dbno_no
    
```

### Frequency table for Vomit

```python
def vomit_freq(c):
    if c=="yes":
            vom_yes=data[data.Vomit=="yes" ]
            vom_yes_yes=vom_yes[vom_yes['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            vom_yes_no=vom_yes[vom_yes['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return vom_yes_yes,vom_yes_no
    if c=="no":
            vom_no=data[data.Vomit=="no" ]
            vom_no_yes=vom_no[vom_no['Disease Z']=="yes"]['Disease Z'].count()/data[data['Disease Z']=="yes"]['Disease Z'].count()
            vom_no_no=vom_no[vom_no['Disease Z']=="no"]['Disease Z'].count()/data[data['Disease Z']=="no"]['Disease Z'].count()
            return vom_no_yes,vom_no_no
    
```

### Finding out Likelihood and Probability Prediction 

```python
def likelihood():
    a=input("enter your status of BP")
    o=input("enter your status of fever")
    b=input("enter your status of diabetes")
    c=input("enter your status of vomit")
    bpy,bpn=bp_freq(a)
    fy,fn=fever_freq(o)
    dy,dn=diabetes_freq(b)
    vy,vn=vomit_freq(c)
    
       
    yes=data[data['Disease Z']=="yes"]['Disease Z'].count()/data.shape[0]  
    yes_prob=bpy*fy*dy*vy*yes
    print("Query|yes = ",yes_prob)
    
    
    no=data[data['Disease Z']=="no"]['Disease Z'].count()/data.shape[0]
    no_prob=bpn*fn*dn*vn*no
    print("Query|no = ",no_prob)
    
    if yes_prob>no_prob:
        print("Class yes has higher probability than class no =",yes_prob)
    else:
        print("Class no has higher probability than class yes =",no_prob)
    
```
