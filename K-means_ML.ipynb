## Distance , Clustering and Calculation of New Centroid Function

```python
def dist(c1x,c1y,c2x,c2y):
    for i in range(data.shape[0]):
        
        dist1=math.sqrt((c1x-data.X[i])**2 + (c1y-data.Y[i])**2)
        data['Distance1'][i]=dist1
        dist2=math.sqrt((c2x-data.X[i])**2 + (c2y-data.Y[i])**2)
        data['Distance2'][i]=dist2
        
        if data.Distance1[i]<data.Distance2[i] :
            data['Cluster'][i]=1
     
        else:
            data['Cluster'][i]=2
    decimals=pd.Series([3,3],index=['Distance1','Distance2'])
    data.round(decimals)
    new_centroid(c1x,c1y,c2x,c2y)
    
    
def new_centroid(c1x,c1y,c2x,c2y):
    count1,count2=0,0
    newc1x,newc1y=0,0
    newc2x,newc2y=0,0
    for i in range(data.shape[0]):
        if data.Cluster[i]==2:
            count2=count2+1
            newc2x=newc2x+data['X'][i]
            newc2y=newc2y+data['Y'][i]
        elif data.Cluster[i]==1:
            count1=count1+1
            newc1x=newc1x+data['X'][i]
            newc1y=newc1y+data['Y'][i]


    newc1x,newc1y=newc1x/count1,newc1y/count1
    newc2x,newc2y=newc2x/count2,newc2y/count2

    if c1x!=newc1x or c1y!=newc1y or c2x!=newc2x or c2y!=newc2y:
        dist(newc1x,newc1y,newc2x,newc2y)
        
```
